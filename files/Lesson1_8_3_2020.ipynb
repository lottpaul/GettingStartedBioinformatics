# Floating-point Accuracy and Error

When working with floating point numbers be aware that there is a interplay between range and precision.  Meaning that there is a rounding error introduced every time you use them.

**Case 1**: demonstrates when you are using numbers of very different scale.   Because of the precision of the Floating-point number, the larger number may consume the smaller.

````python
1.0 + 0.0000000000000001
````

**Case 2**: When repeating multiple operations, the error introduced by floating point numbers is magnified.

````python
x = 0.01
sum = 0

#Add 0.01 one hundred times to get sum of 1.0
for n in range(0,100):  
   sum = sum + x

print(sum)
````

````python
if sum == 1.0:
  print("Same")
````

````python
expected = 1.0
if (sum == expected):
    print("Yeah, we get the correct answer of 1")
else: 
    print("Wait a second there is some error")
````

**Correcting Last Example**: This last example also illustrates that when we compare floating point numbers, we should never compare them with a simple ==.   We need to account for the error in floating point numbers, so the above should be corrected to something like -

````python
#Get the absolute difference between the two numbers and 
#check if it is less then an allowable error
if (abs(sum-expected) <= 0.0000001):
    print("Yeah, we get the correct answer of 1")
else: 
    print("Wait a second there is some error")
````

# Sets and Dictionaries


## Sets

Sets are an unordered collection of unique elements. Sets are typically used to test if something exists in your data or elimination of duplicates.

Sets allow you to quickly take the Union, Intersection, identify differences and symmetric difference

Sets are defined using curly braces or `set()`.


### Set Functions and Methods

|Function | Description |
|---      |---          |
| len(s) | Return the cardinality of set s. |
| x in s | Test x for membership in s. |
| x not in s | Test x for non-membership in s. |
| s.isdisjoint(other) | Return True if the set has no elements in common with other. Sets are disjoint if and only if their intersection is the empty set. |
| s.issubset(other)<br>set <= other | Test whether every element in the set is in other. |
| set < other | Test whether the set is a proper subset of other, that is, set <= other and set != other. |
| issuperset(other)<br>set >= other | Test whether every element in other is in the set. |
| set > other | Test whether the set is a proper superset of other, that is, set >= other and set != other. |
| s.union(other, ...)<br>set &#124; other &#124; ... | Return a new set with elements from the set and all others. |
| s.intersection(other, ...)<br>set & other & ... | Return a new set with elements common to the set and all others. |
| s.difference(other, ...)<br>set - other - ... | Return a new set with elements in the set that are not in the others. |
| s.symmetric_difference(other)<br>set ^ other | Return a new set with elements in either the set or other but not both. |
| s.copy() | Return a new set with a shallow copy of s. |
| s.update(other, ...)<br>set &#124;= other &#124; ... | Update the set, adding elements from all others. |
| s.intersection_update(other, ...)<br>set &= other & ... | Update the set, keeping only elements found in it and all others. |
| s.difference_update(other, ...)<br>set -= other &#124; ... | Update the set, removing elements found in others. |
| s.symmetric_difference_update(other)<br>set ^= other | Update the set, keeping only elements found in either set, but not in both. |
| s.add(elem) | Add element elem to the set. |
| s.remove(elem) | Remove element elem from the set. Raises KeyError if elem is not contained in the set. |
| s.discard(elem) | Remove element elem from the set if it is present. |
| s.pop() | Remove and return an arbitrary element from the set. Raises KeyError if the set is empty. |
| s.clear() | Remove all elements from the set. |

The subset and equality comparisons do not generalize to a total ordering function. For example, any two nonempty disjoint sets are not equal and are not subsets of each other, so all of the following return False: `a<b, a==b, or a>b`.

Set elements, like dictionary keys, must be hashable.


````python
rna = { 'A', 'G', 'C', 'U', 'a', 'c', 'g', 'u'}
print(rna)
````

````python
rna & dna  #AND (Intersection) - Letters in both nucleotides and dna
````

# Dictionaries
Dictionaries add onto `sets` by assigning a `value` to each unique item - `key`.
Dictionaries in other languages are often called hashes, maps, or associative array.  They are indexed by the `key`.

***Note: Dictionaries are unsorted.   The order you is not the order you may get them out.***

You create dictionaries using curly brackets and separate keys and values using a colon.

    dictionary = {key: value, key: value, ...}

You can access the value by appending square brackets that surround the key to the dictionary variable

    dictionary['key']  #Returns the value of 'key'



### Dictionaries Functions

## Identity vs Equality :  `is` vs '=='
There is a difference between `is` and `==`.  

`is` tests that the identity of the value is the same.

`==` tests that the values are equal to each other.

Note:  For built in types like strings, lists, dict - if x is y, then x == y.  But if x==y it doesn't mean that x is y.

So, be sure to use `==` when you are comparing values and `is` when you are comparing identity


````python
x = 1.01
y = 1.01
if x is y:
    print("Same")
elif x == y:
    print("X is equal to Y")

A = "By"
B = "By"
if A is B:
    print("Same")
elif A == B:
    print("X is equal to Y")

x = 1.01
y = x
if x is y:
    print("Same")
elif x == y:
    print("X is equal to Y")
````

# Functions

Functions are blocks of code that allow you to easily reuse code.  Typically, you try to keep a function simple and performs only one task.

**Note**:  It is much easier to find an error in function that is 10 lines of code than 1000 lines of code.

## Function Syntax

```
def <function name>(<function arguments>):
    """
    Description of the Function
    """
    <code>
    
    return <return value>   # Default is None
```

````python
def square(x=0):
    y = x*x
    print(f"The square of {x} is {y}")
    return

square(2)
````

````python
def mult(x, y=None):
  """
  Mult multiple x and y
  """
  z = x*y
  print(f"The multiple of {x} and {y} is {z}")
  return

mult('X',4)
````

Now, `square` calculates the square and prints it out.  But what if we needed to use the value in a different calculation - it doesn't return it. 

To make the `square` more useful, we need to simplify it and do only one thing - calculate the value and return it.

````python
def square(x):
    return x*x

x = 2
y = square(x)
print(f"The square of {x} is {y}")

area = 3.14 * square(2)
print(area)
````

## Default Arguments
You can define default values to arguments.

```
def <function name>(argument(s), default_argument=default_value):

    return
```


````python
def power(x,exponent=2,integer=False):
    p = x**exponent
    if integer:
        return int(p)
    return p
print(power(2.2,exponent=5))
print(power(2.2,integer=True))
print(power(2.2))
````

````python
a = [(lambda x: x * 2)(x) for x in range(5)] 
print(a)
````
